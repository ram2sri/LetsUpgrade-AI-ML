Accuracy = (TP+TN)/(TP+TN+FP+FN)

```python
Accuracy = (42+32)/(42+32+8+18)*100
Accuracy
```

precision = TP/(TP+FP)

```python
Precision = 42/(42+32)
print(Precision)
```

```python
Recall = 42/(42+18)
Recall
```
